```python
import pandas as pd
import numpy as np

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score

import joblib
```

```python
df = pd.read_csv("Wholesale customers data.csv")
```

```python
df_clean = df.drop(columns=["Channel", "Region"], errors="ignore")

X = df_clean.copy()
```

```python
kmeans = joblib.load("kmeans_wholesale.pkl")
scaler = joblib.load("scaler_wholesale.pkl")
```

```python
X_scaled = scaler.transform(X)
labels_original = kmeans.predict(X_scaled)

baseline_silhouette = silhouette_score(X_scaled, labels_original)
print("Baseline silhouette score:", round(baseline_silhouette, 4))
```

```python
#Remove 20% of data
X_reduced = X.sample(frac=0.8, random_state=42)
```

```python
X_reduced_scaled = scaler.transform(X_reduced)
```

```python
labels_reduced = kmeans.predict(X_reduced_scaled)
```

```python
silhouette_reduced = silhouette_score(X_reduced_scaled, labels_reduced)
print("Silhouette score (20% data removed):", round(silhouette_reduced, 4))
```
